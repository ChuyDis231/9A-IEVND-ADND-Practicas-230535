*12.3 Funciones que mandan a llamar a otras funciones y que regresan el resultado de un procesamiento,* en este caso utilizaremos una función predefinida en Python denominada *replace* que nos permite sustituir letras en una cadena de texto. 

*Definición:*

```python
def sustituir_vocales(texto):
    texto = texto.replace("A","4")
    texto = texto.replace("a","4")

    texto = texto.replace("E","3")
    texto = texto.replace("e","3")

    texto = texto.replace("I","1")
    texto = texto.replace("i","1")

    texto = texto.replace("O","0")
    texto = texto.replace("o","0")

    texto = texto.replace("U","∪")
    texto = texto.replace("u","∪")

    return texto
```

*Llamado:*

```python
print(sustituir_vocales(saludar_nombre("Chuy")))
```

**18. Eliminar los datos de los cuartiles menores**

Con la finalidad de hacer un análisis centrado en los datos tendencia es decir con mayor frecuencia eliminamos los registros que entran en el cuartil menor a 0.05 y mayor a 0.95, utilizando la creación de un nuevo dataframe (df3), recordando que la muestra completa sigue en el df original.

```python
a = df['MILES'].quantile(0.95)
b = df['MILES'].quantile(0.05)
df3 = df[df['MILES']>b]
df3 = df[df['MILES']<a]
df3
```

**19.Cálculo de la Desviación Estándar**

Para entender que tan variables o dispersas son las distancias de los viajes.

```python
df3['STD'] = df.MILES.std()
df3
```

**20. Estimar el valor de Z-Score o Puntuación Z**

Con la finalidad de estimar la dispercción de datos, es decir, que tan variados o diferentes son, utilizamos esta formula estadística para poder continuar con el análisis EDA (Exploratory Data Analysis)

```python
df3['Z'] = (df.MILES-df.MILES.mean())/df.MILES.std()
df3
```

**25. Gráfico de Lugar de Abordaje (START)**
